# Nba Regular Season 2018-19 Data Challenge

Your task will be to take the dataset given, and create an analysis answering the following 10 questions. This project will again test your knowledge of pandas in order to find the answers needed given the data you are presented with.

```python
import pandas as pd
import numpy as np
nba_data = pd.read_csv('nbastats2018-2019.csv')
nba_data
# Name,Height,Weight,Team,Age,Salary,Points,Blocks,Steals,
# Assists,Rebounds,FT%,FTA,FG3%
# FG3A,FG%,FGA,MP,G,PER,OWS,DWS,WS,WS48,USG,BPM,VORP
```

# What was the average age of player in the league?

```python
nba_avg_age = nba_data['Age'].mean()
nba_avg_age.round()
```

# What player scored the most points?

```python
nba_max_points = nba_data.groupby('Name', as_index=True).max()[['Points']]
nba_max_points.sort_values(by=['Points'], ascending=False).head(1)
```

# What player had the most blocks during the season? Was it a post player (F/C)?

```python
blocks = nba_data.sort_values(by=['Blocks'], ascending=False)
blocks[['Name', 'Blocks']].head(1)
```

# What player had the best 3-pt percentage? 

```python
threes = nba_data.sort_values(by=['FG3%'], ascending=False)
threes[['Name', 'FG3%']].head(4)
```

# Who played the most minutes during the season

```python
nba_time = nba_data.sort_values(by=['MP'], ascending=False)
nba_time[['Name', 'MP']].head(2)
```

# What player given their player effiecency rating was the clutchest during the season?

```python
# nba_max_points = nba_data.groupby('Name', as_index=True).max()[['Points']]
# nba_max_points.sort_values(by=['Points'], ascending=False).head(1)
nba_efficiency = nba_data.groupby('Name', as_index=True).max()[['PER']]
nba_efficiency.sort_values(by=['PER'], ascending=False).head(1)
```

# What team had the youngest roster?

```python
nba_team_ages = nba_data.groupby(['Team']).sum()
nba_team_ages = nba_team_ages.sort_values(by=['Age']).head(1)
nba_team_ages[['Age']]
```
